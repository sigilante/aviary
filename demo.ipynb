Aviary evaluates combinatory-logic expressions under **normal-order
(leftmost-outermost) reduction to full normal form**: applications are
left-associative and implicit (`S K K` means `((S K) K)`), and results
are printed with a step count.


## 1. The Mockingbird fondness theorem

**Definition.** A bird `β` is *fond of* a bird `α` if `β α = α (β α)`.

**Theorem** (Smullyan). If a forest contains a Mockingbird (`M a = a a`)
and is combinatorially complete, then every bird has a bird fond of it.

Aviary preloads `Θ` (Turing's fixed-point bird) as an alias for `U U`,
where `U` is Smullyan's Turing Bird (`U a b = b (a a b)`) -- itself
built from nothing but self-application, in the same spirit as `M`.
`Θ` is a *universal* witness: `Θ` is fond of *every* bird, since
unfolding its definition gives exactly the fondness equation.


```combinatory-logic
%whnf Θ f
```

(We used `%whnf` here rather than a plain expression: `f` is a free
variable, so the *full* normal form of `Θ f` would try to keep
normalizing the leftover `U U f` inside the result forever -- same
divergence as `Y f`. Weak head normal form stops right where the
fondness equation becomes visible.)

One step unfolds the alias `Θ -> U U`; the second step fires `U`'s own
rule, `U a b = b (a a b)`, giving `f (U U f)` -- and `U U` **is** `Θ`
(that's its very definition), so this is exactly `f (Θ f)`. `Θ` is fond
of `f`, for any `f` at all -- that's the theorem.

Smullyan's classical *construction* of such witnesses (rather than using
the preloaded `Θ`) uses the Cardinal and Bluebird together with the
Mockingbird. The key lemma: for any bird `y`, `C B y` satisfies
`(C B y) x = x (y x)` for every `x`. Let's see the lemma's first step,
specializing `y := M`:


```combinatory-logic
C B M ▲
```

`C` fired once, giving `B ▲ M` -- an under-applied Bluebird, so this is
already in normal form (`B` needs one more argument to fire). Supplying
one more argument makes the lemma's shape (`x (M x)`) visible; note that
`M z` itself then normalizes further, to `z z`:


```combinatory-logic
C B M ▲ z
```

## 2. `%trace` of the Phoenix

The Phoenix, `Φ a b c d = a (b d) (c d)`, applies both `b` and `c` to a
shared argument `d`, then combines the results with `a` -- like `Ψ`, but
threading the extra argument through instead of pre-composing. Watch it
unfold, contraction by contraction, with `%trace`:


```combinatory-logic
%trace Φ B C K x y
```

## 3. Defining a new combinator

New combinators can be defined in-session by a rewrite rule
(`Name v1 v2 ... -> expr`) or by alias (`Name := expr`). Rules reduce
and basis-expand exactly like built-ins. Let's define a small
permuting bird, `🟢`:


```combinatory-logic
🟢 x y z -> x (z y)
```

```combinatory-logic
🟢 ▲ 🟡 🔵
```

## 4. `%ski` basis expansion

Every combinator's S/K/I form is *derived* by bracket abstraction, not
hand-copied -- including `Q₁` (`Q₁ a b c = a (c b)`), the Quixotic Bird.
`%ski` prints the expansion without reducing it:


```combinatory-logic
%ski Q₁
```

To confirm this derived form is really behaviorally equivalent to `Q₁`
(not just superficially similar), apply both to the same fresh atoms and
reduce:


```combinatory-logic
Q₁ ◆ ● ▲
```
